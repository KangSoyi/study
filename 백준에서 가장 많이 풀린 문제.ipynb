```python
# 2839번 설탕 배달

N = int(input())
bong = []


for i in range(N):
    if (((N-i)%5)+i)%3 ==0:
        bong.append((N-i)//5 + (((N-i)%5)+i)//3)

if len(bong) >= 1:
    print(min(bong))
else:
    print(-1)
```

```python
# 1110번 더하기 사이클

N = int(input())
kai = 0
good = 0

while good != N:
    if N >= 10:
        a = N//10
        b = N - a*10
        c = a + b
        if c >= 10:
            d = c//10
            e = c - d*10
            c = e
        good = b*10 + c
        kai += 1
    else:
        good = N*10 + N
        kai += 1
        
print(kai)
```

```python
# 1110번 더하기 사이클

N = input()
kai = 0
good = 0

if int(N) >= 10:
    while good != int(N):
        try:
            dyu = list(map(int, N.split()))
            c = sum(dyu)
            if c >= 10:
                dyo = list(map(int, str(c).split()))
                c = dyo[1]
            else:
                continue
            good = dyu[1]*10 + c
            kai += 1
        except:
            break
else:
    while good != N:
        try:
            good = N*10 + N
            kai += 1
        except:
            break
        
print(kai)
```
