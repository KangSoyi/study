```python
# 8958번 OX퀴즈

case = int(input())

for i in range(case):
    result = list(map(str, input().split()))
    result = [char for string in result for char in string]
    final = 0
    
    for j in range(len(result)):
        count = 1
        if result[j] == "O":
            for x in range(1, j+1):
                if result[j-x] == "O":
                    count += 1
                elif result[j-x] == "X":
                    break
            final += 1*count
        elif result[j] == "X":
            continue
    
    print(final)
```
